### Regras

### Objetivo
Trabalhar com estado interno do objeto e regras de negócio.

````python
class Aluno:
  def __init__(self, nome: str, matricula: str, progresso: int):
    self.nome=nome
    self._matricula=matricula
    if progresso>100:
      raise ValueError ('Era esperado ValueError ao ultrapassar 100%')
    if progresso<0:
      raise ValueError ('Era esperado ValueError para progresso inicial inválido')
    self.__progresso=progresso

  def exibir(self):
    return self.nome

  def get_matricula(self):
    return self._matricula

  def get_progresso(self):
    return self.__progresso

  def atualizar_progresso(self, valor):
    Progressoatual=self.get_progresso()
    Progressoatual+=valor
    if Progressoatual>100:
      raise ValueError ('Era esperado ValueError ao ultrapassar 100%')
    if Progressoatual<0:
      raise ValueError ('Era esperado ValueError para progresso inicial inválido')
    self.__progresso=Progressoatual
````

## Questão 17 – Mini sistema completo

Implemente um mini sistema com as classes:

### Funcionalidades
- matrícula
- progresso

- histórico

### Objetivo
Integrar os conceitos trabalhados ao longo da lista.

````python
class Curso:
    def __init__ (self, titulo: str, carga_horaria: float, preco: float):
        self.titulo=titulo
        if carga_horaria <0:
          raise ValueError ('Era esperado ValueError para carga horária negativa')
        self._carga_horaria=carga_horaria
        if preco <0:
          raise ValueError ('Era esperado ValueError para preço negativo')
        self.__preco=preco

    def __str__(self):
      return f'Título: {self.titulo}, Carga Horária: {self._carga_horaria}, Preço {self.__preco}'

    def exibir_dados(self):
        return self.titulo, self._carga_horaria, self.__preco

    def get_preco(self):
      return self.__preco

    def get_cargahoraria(self):
      return self._carga_horaria

    def aplicar_desconto(self, percentual: float):
        if percentual>50:
          raise ValueError ('Era esperado ValueError para desconto maior que 50%')
        else:
          self.__preco-=self.__preco*(percentual/100)

class Aluno:
  def __init__(self, nome: str, matricula, progresso, historico: list = None):
    self.nome=nome
    self._matricula=matricula
    self.__progresso=progresso
    self.__historico=historico if historico is not None else []

  def __str__(self):
    return f'Nome: {self.nome}, Matricula: {self._matricula}, Progresso: {self.__progresso}, Histórico: {self.__historico}'

  def exibir(self):
    return self.nome
  def get_matricula(self):
    return self._matricula
  def get_progresso(self):
    return self.__progresso
  def get_historico(self):
    return self.__historico

  def atualizar_progresso(self, valor):
    Progressoatual=self.get_progresso()
    Progressoatual+=valor
    if Progressoatual>100:
      raise ValueError ('Era esperado ValueError ao ultrapassar 100%')
    if Progressoatual<0:
      raise ValueError ('Era esperado ValueError para progresso inicial inválido')
    self.__progresso=Progressoatual
    self.__historico.append(self.__progresso)

class Inscricao:
    def __init__(self, aluno, curso):
      self.aluno=aluno
      self.curso=curso

    def __str__(self):
      return f'Titulo: {self.aluno}, Carga Horária: {self.curso}'

    def registrar_acesso(self, horas_estudadas):
      cargahoraria=self.curso.get_cargahoraria()
      valor=horas_estudadas
      Progresso_Curso= (horas_estudadas/cargahoraria)*100
      self.aluno.atualizar_progresso(Progresso_Curso)

    def esta_concluida(self):
      if self.aluno.get_progresso() == 100:
        return True
      return False
    def emitir_certificado(self, formato="PDF", assinatura=True):
      self.formato=formato
      self.assinatura=assinatura
      return f'Formato: {formato}, Assinatura: {assinatura}'
````

````python
# Testes da Questão 17
curso = Curso("Python OO", 40, 200)
aluno = Aluno("Marcos", "2026001", 0)
inscricao = Inscricao(aluno, curso)

inscricao.registrar_acesso(20)
assert aluno.get_progresso() == 50

inscricao.registrar_acesso(20)
assert aluno.get_progresso() == 100
assert inscricao.esta_concluida() is True

cert = inscricao.emitir_certificado()
assert isinstance(cert, str)

hist = aluno.get_historico()
assert isinstance(hist, list)
assert len(hist) >= 2

print("✅ Questão 17 correta")
````
